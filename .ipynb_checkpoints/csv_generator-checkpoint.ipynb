```python
working_directory = os.getcwd() + '/90kDICT32px'
working_directory
```

#Â All Dataset

```python
paths = []
with open('./90kDICT32px/annotation_train.txt', 'r') as f:
    for l in f:
        components = l.split(' ')
        path = components[0][1:]
        label = path.split('_')[1]
        paths.append([label, path])
```

```python
# all_train_df = pd.DataFrame(paths, columns = ['label','path'])
all_train_df = pd.read_csv('all_train.csv')
```

```python
all_train_df.head()
```

```python
all_train_df.drop(columns = ['Unnamed: 0'], inplace = True)
all_train_df.head()
```

```python
file_path = working_directory + all_train_df['path'][0]
img = plt.imread(file_path, format = 'CGRAY')
plt.imshow(img, cmap = 'gray')
```

```python
def draw(rows, cols, df):
    _, axs = plt.subplots(rows, cols, figsize = (16,12))
    sample = df.sample(rows * cols)

    for idx, ax in enumerate(axs.flatten()):
        label, path = sample.values[idx]
        img = plt.imread(working_directory + path)
        ax.imshow(img)
        ax.set_title(label)
        ax.axis('off')
```

```python
draw(5, 5, all_train_df)
```
